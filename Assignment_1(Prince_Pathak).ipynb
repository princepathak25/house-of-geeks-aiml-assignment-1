**Question 2-** Write a Python program that calculates the average (mean) of a given list of numerical values without using the built-in sum() function or any other library function that directly computes the total.


```python
def calculate_average(numbers):
    total = 0      # stores the sum of numbers
    count = 0      # stores how many numbers are there

    # loop through each number in the list
    for num in numbers:
        total += num
        count += 1

    # check to avoid division by zero
    if count == 0:
        return 0

    # return the average value
    return total / count


# take input from the user
data = eval(input("Enter a list of numbers (e.g., [1, 2, 3]): "))

# calculate the average
avg = calculate_average(data)

# print the result
print("The average is:", avg)
```

**Question 3-** Write a Python program to generate and display the Fibonacci series up to a given number of terms.

```python
def fibonacci_series(n):
    fib = []        # list to store Fibonacci numbers
    a, b = 0, 1     # first two Fibonacci values

    for i in range(n):
        fib.append(a)
        a, b = b, a + b

    return fib

terms = int(input("Enter number of terms: "))
result = fibonacci_series(terms)
print("Fibonacci series:", result)
```

**Question 4-** Write a Python program to print numbers from 1 to N following the rules below:

*   Print ‚ÄúFizz‚Äù if the number is divisible by 3

*   Print ‚ÄúBuzz‚Äù if the number is divisible by 5
*   Print ‚ÄúFizzBuzz‚Äù if the number is divisible by both 3 and 5


*   Otherwise, print the number itself




```python
n = int(input("Enter the value of N: "))

for i in range(1, n+1):

    # number divisible by both 3 and 5
    if i % 3 == 0 and i % 5 == 0:
        print("FizzBuzz")

    # number divisible by only 3
    elif i % 3 == 0:
        print("Fizz")

    # number divisible by only 5
    elif i % 5 == 0:
        print("Buzz")

    else:
        print(i)
```

**Question 5-** Write a Python program to calculate the mean, median, and mode of a given list of
numerical values.

The program should compute each statistical measure manually, without using any built-in
statistical functions or external libraries.

Requirements:
 1. Accept a list of integers or floating-point numbers.
 2. Sort the list to calculate the median.
 3. Count occurrences of each value to determine the mode.
 4. Display the mean, median, and mode clearly.
 5. Handle edge cases such as:

    ‚Ä¢ Empty list

    ‚Ä¢ Multiple modes


```python
def calculate_mean(numbers):
    # if list is empty, mean not possible
    if len(numbers) == 0:
        return None

    return sum(numbers) / len(numbers)

def calculate_median(numbers):
    # check empty list
    if len(numbers) == 0:
        return None

    nums_sorted = sorted(numbers)
    n = len(nums_sorted)
    mid = n // 2

    # even number of elements
    if n % 2 == 0:
        return (nums_sorted[mid - 1] + nums_sorted[mid]) / 2

    # odd number of elements
    return nums_sorted[mid]


def calculate_mode(numbers):
    # empty list case
    if len(numbers) == 0:
        return None

    count = {}
    # count frequency of each number
    for num in numbers:
        if num in count:
            count[num] += 1
        else:
            count[num] = 1

    max_freq = max(count.values())
    modes = []

    for key in count:
        if count[key] == max_freq:
            modes.append(key)

    # if all values occur once
    if max_freq == 1:
        return None

    return modes

def main():
    data = eval(input("Enter a list of numbers: "))

    mean = calculate_mean(data)
    median = calculate_median(data)
    mode = calculate_mode(data)

    print("Mean:", mean)
    print("Median:", median)

    if mode is None:
        print("Mode: No mode")
    else:
        print("Mode:", end=" ")
        for m in mode:
            print(m, end=" ")
        print()
main()
```

**Question 6-** Write a Python program to calculate the standard deviation of a given list of numerical values without using any built-in statistical or numerical libraries.


```python
def calculate_standard_deviation(data):
    if len(data) == 0:
        return 0.0

    # Calculate the mean
    mean = sum(data) / len(data)

    # Calculate the variance
    variance = sum((x - mean) ** 2 for x in data) / len(data)

    # Calculate the standard deviation
    standard_deviation = variance ** 0.5

    return standard_deviation

data = eval(input("Enter a list of numbers: "))
std_dev = calculate_standard_deviation(data)
print("The standard deviation of the data set is:",std_dev)
```

**Question 7-** Write a Python program to perform matrix addition for two matrices of the same dimensions using pure Python, without using any external libraries such as NumPy.

Defination:

                       ùë® = [ùíÇùíäùíã] and ùë© = [ùíÉùíäùíã]

Then the sum matrix ùë™ = ùë® + ùë©is defined as:

                       ùíÑùíäùíã = ùíÇùíäùíã + ùíÉùíäùíã

Program Requirements
1. Accept two matrices as input in the form of nested lists.
2. Ensure both matrices have the same number of rows and columns.
3. Add corresponding elements of the matrices.
4. Store the result in a new matrix.
5. Display the resulting matrix in proper matrix format.
6. Handle invalid cases where matrix dimensions do not match.

```python
def add_matrices(A, B):

    # Check if number of rows are equal
    if len(A) != len(B):
        print("Matrix addition not possible (row size mismatch)")
        return

    # Check if number of columns are equal
    if len(A[0]) != len(B[0]):
        print("Matrix addition not possible (column size mismatch)")
        return

    # Create an empty matrix to store the result
    result = []

    for i in range(len(A)):
        row = []
        for j in range(len(A[0])):
            row.append(A[i][j] + B[i][j])
        result.append(row)

    # Display the result matrix
    print("Resultant Matrix:")
    for r in result:
        print(r)

A = eval(input("Enter first matrix (as nested list): "))
B = eval(input("Enter second matrix (as nested list): "))
add_matrices(A, B)
```

**Question 8-** Write a Python program to calculate Mean Absolute Error (MAE) and Mean Squared
Error (MSE) between actual values and predicted values without using any built-in statistical
libraries.

Definitions:

Where:

‚Ä¢ ùíöùíä= actual value

‚Ä¢ ùíöÃÇùíä= predicted value

‚Ä¢ ùíè= number of observations

Program Requirements
1. Accept two lists:



    *   One containing actual values
    *   One containing predicted values



2. Ensure both lists are of equal length.

3. Calculate Both MSE and MAE

```python
def find_mae(actual, predicted):

    total = 0
    n = len(actual)
    for i in range(n):
        diff = actual[i] - predicted[i]
        if diff < 0:
            diff = -diff
        total = total + diff
    return total / n

# Function to calculate Mean Squared Error
def find_mse(actual, predicted):

    total = 0
    n = len(actual)

    for i in range(n):
        diff = actual[i] - predicted[i]
        total = total + diff * diff

    return total / n

n = int(input("Enter number of values: "))

actual_values = []
predicted_values = []

# Input actual values
print("Enter actual values:")
for i in range(n):
    value = float(input())
    actual_values.append(value)

# Input predicted values
print("Enter predicted values:")
for i in range(n):
    value = float(input())
    predicted_values.append(value)

# Calculate MAE and MSE
mae = find_mae(actual_values, predicted_values)
mse = find_mse(actual_values, predicted_values)

# Display results
print("Mean Absolute Error =", mae)
print("Mean Squared Error =", mse)
```

**Question 9-** Write a Python program to calculate the determinant of a square matrix.

Program Requirements
1. Accept a square matrix (2√ó2 or 3√ó3).
2. Validate that the matrix is square.
3. Use the determinant formula:
4. Display the determinant value.
5. Print an error message for non-square matrices.

```python
def determinant(m):

    n = len(m)
    # Check whether the matrix is square
    for row in m:
        if len(row) != n:
            print("Matrix is not square")
            return
    if n == 2:
        print("Determinant =", m[0][0]*m[1][1] - m[0][1]*m[1][0])
    elif n == 3:
        print("Determinant =",
              m[0][0]*(m[1][1]*m[2][2] - m[1][2]*m[2][1])
            - m[0][1]*(m[1][0]*m[2][2] - m[1][2]*m[2][0])
            + m[0][2]*(m[1][0]*m[2][1] - m[1][1]*m[2][0]))

    else:
        print("Only 2x2 or 3x3 matrices are allowed")

n = int(input("Enter matrix order (2 or 3): "))
matrix = []
print("Enter elements row-wise:")
# Read matrix elements
for i in range(n):
    row = []
    print("Row", i + 1)
    for j in range(n):
        row.append(int(input()))
    matrix.append(row)
# Call the function to find determinant
determinant(matrix)
```
